```python
import random
```

Person class

```python
class Person:
    def __init__(self, name, length):
        self.name = name
        self.length = length
        self.bits = []
        self.bases = []
        self.states = []
    
    def randomiseBits(self):
        for i in range(self.length):
            self.bits.append(random.choice([0, 1]))
        

    def randomiseBases(self):
        for i in range(self.length):
            self.bases.append(random.choice(["X", "Z"]))

    def deduceStates(self):
        states = ["|->", "|1>", "|+>", "|0>"]
        
        for i in range(self.length):
            if(self.bits[i] == 1):
                if(self.bases[i] == "X"):
                    self.states.append(states[0])
                else:
                    self.states.append(states[1])
            else:
                if(self.bases[i] == "X"):
                    self.states.append(states[2])
                else:
                    self.states.append(states[3])   

    def deduceBits(self, array_bits, array_bases):
         bits = [0, 1]
         for i in range(self.length):
            if(self.bases[i] == array_bases[i]):
                self.bits.append(array_bits[i]) # Step 5: If the bases match then so must the bits, so it is measured
            else:
                self.bits.append(random.choice(bits)) 

    def display(self):
        print(self.bits)
        print(self.bases)
        print(self.states)
```

Alice

```python
length = 100
Alice = Person("Alice", length)
Alice.randomiseBits()
Alice.randomiseBases()
Alice.deduceStates()
Alice.display()
```

Eve

```python
# Eve has access to Alice's bases
Eve = Person("Eve", length)
Eve.randomiseBases()
Eve.deduceBits(Alice.bits, Alice.bases)
Eve.deduceStates()
Eve.display()
```

Bob

```python
# Bob has access to Eve's bases
Bob = Person("Bob", length)
Bob.randomiseBases()
Bob.deduceBits(Eve.bits, Eve.bases)
Bob.deduceStates()
Bob.display()
```

One Time Pad

```python
otp_alice= []
otp_bob= []

for i in range(length):
  if(Alice.bases[i] == Bob.bases[i]):
    otp_alice.append(Alice.bits[i])
    otp_bob.append(Bob.bits[i])

print("For Alice: ", otp_alice)
print("For Bob: ", otp_bob)
```

Deducing Error Rate

```python
Match_count  = 0
Sub_length = length//4 # Comparing first one fourth of one time pad
for i in range(Sub_length):
  if(otp_alice[i] == otp_bob[i]):
    Match_count += 1

print(Match_count)
Error_rate = 100 - ((Match_count/Sub_length)*100)
print("Error Rate: ", Error_rate, "%")
```
