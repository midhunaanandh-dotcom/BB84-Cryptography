Alice

```python
import random

length = 50
bits = [0, 1]
bases = ["X", "Z"]
states = ["|->", "|1>", "|+>", "|0>"]

Alice_bits = []
Alice_bases = []
Alice_states = []

# Step 1: Creating a random string of bits and bases
for i in range(length):
  Alice_bits.append(random.choice(bits))
  Alice_bases.append(random.choice(bases))

# Step 2: Deducing the states of each bit according to quantum physics
for i in range(length):
  if(Alice_bits[i] == 1):
    if(Alice_bases[i] == "X"):
      Alice_states.append(states[0])
    else:
      Alice_states.append(states[1])
  else:
    if(Alice_bases[i] == "X"):
      Alice_states.append(states[2])
    else:
      Alice_states.append(states[3])

print(Alice_bits)
print("Alice Bases: ", Alice_bases)
print(Alice_states)
```

Bob

```python
Bob_bits = []
Bob_bases = []

# Step 3: Creating a random list of bases
for i in range(length):
  Bob_bases.append(random.choice(bases))

# Step 4: Remember that Bob only has access to Alice's bases and states(which is not measured yet)
for i in range(length):
  if(Bob_bases[i] == Alice_bases[i]):
    Bob_bits.append(Alice_bits[i]) # Step 5: If the bases match then so must the bits, so it is measured
  else:
    Bob_bits.append(random.choice(bits)) # Step 6: If the bases don't match then well, Bob can still measure the qubit
                                         # but through a wrong "basis", collapsing it into a state of 50/50 chance of ebing a 0 or 1

print("Bob Bases: ", Bob_bases)
print(Bob_bits)
```

Deducing One Time Pad

```python
# Step 7: Alice and Bob reveal their bases publicly

otp_alice= []
otp_bob= []

for i in range(length):
  if(Alice_bases[i] == Bob_bases[i]):
    otp_alice.append(Alice_bits[i])
    otp_bob.append(Bob_bits[i])

print("For Alice: ", otp_alice)
print("For Bob: ", otp_bob)
```

Detecting Error Rate

```python
Match_count  = 0
Sub_length = len(otp_alice)//4 # Comparing first one fourth of one time pad
for i in range(Sub_length):
  if(otp_alice[i] == otp_bob[i]):
    Match_count += 1

print(Match_count)
Error_rate = 100 - ((Match_count/Sub_length)*100)
print("Error Rate: ", Error_rate, "%")
```
